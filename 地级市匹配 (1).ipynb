```python
import pandas as pd
import numpy as np
import time
import os
from tqdm import tqdm
```

```python
def get_province_on_the_road(data_latitude,data_longitude,city_latitude,city_longitude,VIN_num,city_name,province_name):
    province=[]
    city = []
    chassis_num = []
    result = pd.DataFrame()
    with tqdm(total=len(data_latitude)) as pbar:
        for i in range(len(data_latitude)):
            difference=abs(data_latitude[i]-city_latitude)+abs(data_longitude[i]-city_longitude)
            Near_city=np.argpartition(difference,1)
            chassis_number = VIN_num[i][-8:]
            if difference[Near_city[0]] > 0.3:
                province.append('未知')
                city.append('未知')
            else:
                province.append(province_name[Near_city[0]])
                city.append(city_name[Near_city[0]])
            chassis_num.append(chassis_number)
            pbar.update(1)
        result.insert(0,'底盘号',chassis_num)
        result.insert(1,'省份',province)
        result.insert(2,'地级市',city)
        result = result.drop_duplicates()
        return result
```

```python
data_dir = r'D:\原型设计\新能源中重卡充电习惯分析\车辆数据'
city_dir=r'D:\Users\liyy3\Desktop\城市经纬度表.xlsx'
result = pd.DataFrame()
for filename in os.listdir(data_dir):
    file_dir = os.path.join(data_dir,filename)
    data = pd.read_csv(file_dir)
    city=pd.read_excel(city_dir)[['经度','纬度','省份','地级市']]
    data_longtitude,data_latitude=data['经度'].astype(float),data['纬度'].astype(float)
    VIN_num = data['VIN']
    temp=get_province_on_the_road(data_latitude.tolist(),data_longtitude.tolist(),city['纬度'],city['经度'],VIN_num.tolist(),city['地级市'],city['省份'])
    result = result.append(temp)
# print(result)
result.to_excel(r'D:\Users\liyy3\Desktop\结果数据\省份城市匹配.xlsx',index = False)
```
